```python
reviewed_df['Analysis'] = df.sentiment
reviewed_df
```

```python
X = reviewed_df.drop('Analysis', axis=1)
y = reviewed_df.Analysis
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LogisticRegression(max_iter=1000).fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```
